### Kerr

```python
g = MU.metric()

rho2 = r^2 + (a*cos(th))^2
Delta = r^2 -2*M*r + a^2

g[0,0] = -(1-2*M*r/rho2)
g[0,3] = -2*a*M*r*sin(th)^2/rho2
g[1,1] = rho2/Delta 
g[2,2] = rho2
g[3,3] = (r^2 + a^2 + 2*M*r*(a*sin(th))^2/rho2)*sin(th)^2
g.display()
```
